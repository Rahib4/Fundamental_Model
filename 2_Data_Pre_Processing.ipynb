<h1>** Quarterly Data Cleaning **<h1>

```python
## Quarterly Data Cleaning
df_qtrly_sales = pd.read_csv('df_qtrly_sales.csv', low_memory=False)
df_qtrly_sales.head(4)
```

```python
df_qtrly_sales.dtypes
```

```python
# Removing % from all the columns
# Converting them into numeric, They all are object because of mixed data type.
df_qtrly_sales = df_qtrly_sales.replace(',','',regex = True)
df_qtrly_sales = df_qtrly_sales.replace('%','',regex = True)
df_qtrly_sales.head(5)
```

```python
df_qtrly_sales.columns
```

```python

df_qtrly_sales.iloc[:,2:] = df_qtrly_sales.iloc[:,2:].apply(pd.to_numeric)
df_qtrly_sales
```

```python
# changing the names of the columns.. putting year first and month after
col_names = df_qtrly_sales.columns.tolist()
col_name_new = []
for col in col_names:
    try:
        new_col = str(col.split('_')[1]) + str('_') + str(col.split('_')[0])
    except:
        new_col = col
    #print(col,'New name - ',new_col)
    col_name_new.append(new_col)

print(col_names)
print(col_name_new)

col_names_old = df_qtrly_sales.columns
df_qtrly_sales.columns = col_name_new
```

```python
# Removing extra years, keeping data only after 2018 Dec 
df_qtrly_sales = df_qtrly_sales[['Tikr', 'KPI', '2018_Dec',
                              '2019_Mar', '2019_Jun', '2019_Sep', '2019_Dec', '2020_Mar', '2020_Jun', '2020_Sep', '2020_Dec',
                              '2021_Mar', '2021_Jun', '2021_Sep']]
df_qtrly_sales.columns = ['Tikr', 'KPI', '2018_Q4',
                              '2019_Q1', '2019_Q2', '2019_Q3', '2019_Q4', '2020_Q1', '2020_Q2', '2020_Q3', '2020_Q4',
                              '2021_Q1', '2021_Q2', '2021_Q3']                
df_qtrly_sales.head(5)
```

```python
# Pivoting the data.. from KPI column to individual columns with quarter
df = df_qtrly_sales
# Pivotted the pivot
pivot_columns = list(set(df.columns) - set(['KPI','Tikr']))  
df_pivot = pd.pivot_table(df, index=['Tikr'], columns=['KPI'], values=pivot_columns, aggfunc='sum')
# Replacing spaces from the column names and joining the two layer of column names into one.
df_pivot.columns =[str(s2) + str('_') + s1  for (s1,s2) in df_pivot.columns.tolist()]
df_pivot.reset_index(inplace=True)
df_pivot.columns = df_pivot.columns.str.replace('+', '')
df_pivot.columns = df_pivot.columns.str.replace(' ', '_')
# Reordering columns by name so that all similar column get together
df_pivot = df_pivot.reindex(sorted(df_pivot.columns), axis=1)
df_pivot.head(4)
```

```python
# Plan: 
# Test/Train    : I will use data till 2019 as training and data post 2019 as test. [1250 (Train) + (400 Test)]. This is assuming that the Covid impact has only increased the return % but the overall ranking of returns will remain the same.
# Model         : XgBoost + LGBM + RandomForest (dont have sufficient enough data for Neural networks)
# X variable    : Company Financials till 2019.
# Y variable    : last 2 year Return % (2021 vs 2019)- Minus the return from index (so that some compaies which are not performing at par with index will give negative returns, otherwise all companies will give positive returns because of time)
# Results       : We will be categorizing compaies based on binary classification into 10 buckets.. and ideally all higher ranked compaies should fall in the same bracket on the basis of 2020 returns

# Different features that I am planning to create on top of Quarterly and Annual data
#1. QoQ or YoY Percentage change
#2. Average of all these years
#3. CAGR of Sales
#4. Embeddings for sales and revenue etc.
#5. Company Features - 
#   a. MidCap/SmallCap/BlueChip
#   b. Tenure of a company
#   c. Number of shares
#   d. Market cap in 2019
#   e. Company Category/Industry
#   f. Time series cluster - Slowly increasing + Declining + Sharp Rising + Sales increasing + Profit Increasing 

# Future Plan
# 1. Add Cash Flow and holdings data
# 2. Add BSE data (4000 more companies) data
# 3. Add Volumes data
# 4. Add technical indicators like RSI + RSI 200 + RSI 50 etc.
# 5. Add metrics like Petrol + Commodity rates + Inflation rate + India GDP + GDP growth rate %
# 6. Sector Baseline comparisons
# 7.
```

```python
# Variable 1    : Sales
matching = [s for s in df_pivot.columns if "Sales\xa0" in s]        # Getting all the columns containing 'sales/profit etc'
matching.append('Tikr')                                             # Added tikr column
matching = matching[-1:] + matching[:-1]                            # Moving Tikr column from bottom to top of the list
#print(matching)
df_sales = df_pivot[matching]
#print(df_sales.head(4))
df_sales_col = df_sales.columns
df_sales_col = [i.replace('\xa0','') for i in df_sales_col]         # Getting rid of special characters 
print(df_sales_col)
print('Number of cols in df - ',len(df_sales_col))
df_sales.columns = df_sales_col
```

```python
df_sales.columns
```

```python
# disable chained assignments
pd.options.mode.chained_assignment = None 

def percentage_change(col1,col2):
    try:
        out = ((col2 - col1) / col1) * 100
    except:
        out = np.nan
    return out
    #return ((df_sales.loc[:,col2] - df_sales.loc[:,col1])/df_sales.loc[:,col1]) * 100

quarter_cols = [['2019_Q1_QoQ','2018_Q4','2019_Q1'],
             ['2019_Q2_QoQ','2019_Q1','2019_Q2'],
             ['2019_Q3_QoQ','2019_Q2','2019_Q3'],
             ['2019_Q4_QoQ','2019_Q3','2019_Q4'],
             ['2020_Q1_QoQ','2019_Q4','2020_Q1'],
             ['2020_Q2_QoQ','2020_Q1','2020_Q2'],
             ['2020_Q3_QoQ','2020_Q2','2020_Q3'],
             ['2020_Q4_QoQ','2020_Q3','2020_Q4'],
             ['2021_Q1_QoQ','2020_Q4','2021_Q1'],
             ['2021_Q2_QoQ','2021_Q1','2021_Q2'],
             ['2021_Q3_QoQ','2021_Q2','2021_Q3']]

def calc_percentage_diff(df_var,var,col_out,col1,col2):
    col1 = str(var) + str('_') + str(col1)
    col2 = str(var) + str('_') + str(col2)
    col_out = str(var) + str('_') + str(col_out)
    #print(col_out,col1,col2)
    df_var[col_out] = percentage_change(df_var[col1],df_var[col2])
    #print(df_sales[col_out])

var = 'Sales'
#vars = ['Sales','Expenses','','']
for combination in quarter_cols:
    var_out = combination[0]
    var1 = combination[1]
    var2 = combination[2]
    calc_percentage_diff(df_sales,var,var_out,var1,var2)
```

```python
#Throught this loop we will process multiple colmns at once
# We will clean the variables and also calculate the QoQ for each variable

variables = ['Sales','Revenue','Expenses','Net_Profit','Operating_Profit','EPS_in_Rs','Tax_','OPM_']
   
df_quarter = pd.DataFrame()
for var in variables:
    print(var)
    #matching = [s for s in df_pivot.columns if "Sales\xa0" in s]        # Getting all the columns containing 'sales/profit etc'
    matching = [s for s in df_pivot.columns if var in s]        # Getting all the columns containing 'sales/profit etc'
    matching.append('Tikr')                                             # Added tikr column
    matching = matching[-1:] + matching[:-1]                            # Moving Tikr column from bottom to top of the list
    #print(matching)
    df_var = df_pivot[matching]
    #print(df_sales.head(4))
    df_var_col = df_var.columns
    df_var_col = [i.replace('\xa0','') for i in df_var_col]         # Getting rid of special characters 
    print(df_var_col)
    print('Number of cols in df - ',len(df_var_col))
    df_var.columns = df_var_col

    # Calculating QoQ calculations over all available quarter columns
    # List of output and input columns is in quarter_cols
    # Updating each QoQ column 1 by 1
    for combination in quarter_cols:
        try:    
            var_out = combination[0]
            var1 = combination[1]
            var2 = combination[2]
            calc_percentage_diff(df_var,var,var_out,var1,var2)
        except Exception as e:
            print('Error - ',e)
    df_quarter = pd.concat([df_quarter.reset_index(drop=True), df_var], axis=1)
```
